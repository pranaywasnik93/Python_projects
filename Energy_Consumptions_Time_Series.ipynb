```python
## finding monthly or weekly pattern

from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
df_scaled = std_scaler.fit_transform(df.to_numpy())             #array
df_scaled

energy_df = pd.DataFrame(df_scaled, columns=df.columns, index=df.index)         #array to pandas df
energy_df.head(3)
```

```python
energy_df = energy_df.reset_index()
energy_df.head(3)
```

```python
## lets creat a new column for month from Time column

energy_df["month_no_of_year"] = energy_df["Time"].dt.month.astype(str).str.zfill(2) + "-" + energy_df["Time"].dt.month_name().str[:3]

energy_df["day"] = energy_df["Time"].dt.date

energy_df["hour"] = energy_df["Time"].dt.hour

# Extracting the week number and adding it as a new column
energy_df["week_no"] = energy_df["Time"].dt.isocalendar().week

energy_df["day_of_week"] = energy_df["Time"].dt.weekday.astype(str) + " " + energy_df["Time"].dt.day_name().str[:3]    

energy_df.head(3)
```

```python
cols =[i for i in df.columns if i.startswith("Energy_")]
cols.append("month_no_of_year")
cols

_monthly = energy_df[cols].melt(id_vars="month_no_of_year")
_monthly

monthly = _monthly.rename(columns={"value":"avg_load"})[["month_no_of_year", "avg_load"]]
monthly

avg_monthly = monthly.groupby("month_no_of_year").mean()
avg_monthly

fig, ax = plt.subplots(figsize=(15,5))
av_m = sns.lineplot(avg_monthly, palette=["orange"])


std_monthly = _monthly.rename(columns = {"value":"std_load"})[["month_no_of_year", "std_load"]].groupby("month_no_of_year").std()

fig, ax = plt.subplots(figsize=(15,5))
av_m = sns.lineplot(avg_monthly, palette=["orange"])
sns.lineplot(std_monthly, palette=["green"])
```
