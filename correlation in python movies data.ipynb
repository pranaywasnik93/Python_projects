```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  # adjusts the configuration of the plot we will create

#read in the data

df = pd.read_csv(r"C:\pranay practice\Projects\Python projects\movies.csv")
```

```python
#lets look at data

df.head()
```

```python
#looking for missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print("{} : {}%".format(col,pct_missing))


# another easy method to find null without for loop
#df.isnull().sum() 
```

```python
#dropping the null values

df1 = df.dropna().copy()
```

```python
print(df.shape[0])
df1.shape[0]
```

```python
#checking data types
df1.dtypes
```

```python
#changing data type

df1['budget'] = df1['budget'].astype('int64')
df1['gross'] = df1['gross'].astype('int64')
```

```python
df1
```

```python
#if we have to chnge the data type of original df with null values then 

#df['budget'] =pd.to_numeric(df['budget'],errors = 'coerce').fillna(0).astype('int64')     with o or with mean,median etc
```

```python
# the year in released column and year column dont match,lets create new column using released col

df1['new_year'] = df1['released'].str.extract(r'(\d{4})')
```

```python
df1.head()
```

```python
df1 = df1.sort_values(by = "gross" , inplace = False , ascending = False)
df1.head()
```

```python
df1_numerized = df1.copy()

for col_name in df1_numerized.columns :
    if (df1_numerized[col_name].dtype == 'object'):
        df1_numerized[col_name] = df1_numerized[col_name].astype('category')
        df1_numerized[col_name] = df1_numerized[col_name].cat.codes

df1_numerized.head()
    
    
```

```python
correlation_metrix = df1_numerized.corr(method = "pearson" , numeric_only=True)

sns.heatmap(correlation_metrix,annot = True)
plt.title("correlation metrix for numeric features")
plt.xlabel("movie features")
plt.ylabel("movie features")
plt.show()
```
